```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoded_df = df
```

```python
## Encoding person_education
education_order = ['High School', 'Associate', 'Bachelor', 'Master', 'Doctorate']
encoder = OrdinalEncoder(categories=[education_order])
encoded_df['Education_encoded'] = encoder.fit_transform(df[['person_education']]).astype(int)
```

```python
## Encoding person_gender

gender_order = ['female','male']
encoder = OrdinalEncoder(categories=[gender_order])
encoded_df['gender'] = encoder.fit_transform(df[['person_gender']]).astype(int)
```

```python
## Encoding person_home_owenership types

ownership_order = ['RENT','OWN','MORTGAGE','OTHER']
encoder = OrdinalEncoder(categories=[ownership_order])
encoded_df['home_ownership_encoded'] = encoder.fit_transform(df[['person_home_ownership']]).astype(int)
```

```python
## Encoding loan_intent

loan_intent_types = ['PERSONAL', 'EDUCATION', 'MEDICAL', 'VENTURE', 'HOMEIMPROVEMENT','DEBTCONSOLIDATION']
encoder = OrdinalEncoder(categories=[loan_intent_type])
encoded_df['loan__encoded'] = encoder.fit_transform(df[['person_home_ownership']]).astype(int)
```

```python
## Encoding previous_loan_defaults

previous_loans = ['No', 'Yes']
encoder = OrdinalEncoder(categories=[previous_loans])
df['previous_loans'] = encoder.fit_transform(df[['previous_loan_defaults_on_file']]).astype(int)
```

```python
x = df[['person_age','person_income','person_emp_exp','loan_amnt','loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length','credit_score','Education_encoded', 'gender', 'home_ownership_encoded','loan_intent_encoded','previous_loans']]
y = df['loan_status']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(f"Accuracy score - {score:.2f} ")
cm = confusion_matrix(y_test,y_pred)
print("Confusion matrix - \n",cm)
print(classification_report(y_test,y_pred))
```

# Hyperparameter Tuning using GridSeachCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', None], # Norms
    'C': [0.01, 0.1, 1, 10, 100],        # Regularization strength
    'solver': ['newton_cg','bfgs','liblinear','sag', 'saga'],     # Solvers supporting 'l1' or 'elasticnet'
}
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model,cv=cv,scoring='accuracy',param_grid=param_grid)
```
